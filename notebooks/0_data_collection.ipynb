# DATA COLLECTION

## HotPotQA dataset

```python
import pandas as pd
import numpy as np
import os
import dotenv
from huggingface_hub import login
from datasets import load_dataset
```

```python
# Load environment variables
dotenv.load_dotenv()
HUGGINGFACEHUB_API_TOKEN = os.getenv("HUGGINGFACEHUB_API_TOKEN")
```

```python
# Huggung Face login
login(token=HUGGINGFACEHUB_API_TOKEN)
```

```python
# load HotpotQA dataset chia batch
dataset = load_dataset("hotpot_qa", "distractor", streaming=True, batch_size=1000)
```

```python
dataset["train"][0]
```

```python
# turn to pandas dataframe
train = pd.DataFrame(dataset["train"])
valid = pd.DataFrame(dataset["validation"])
```

```python
# check the folder data/hotpot_qa2/ exists, if not create it
if not os.path.exists("../data/hotpot_qa/"):
    os.makedirs("../data/hotpot_qa/")
```

```python
valid.to_csv("../data/hotpot_qa/valid.csv", index=False, encoding='utf-8-sig')
train.to_csv("../data/hotpot_qa/train.csv", index=False, encoding='utf-8-sig')
```
